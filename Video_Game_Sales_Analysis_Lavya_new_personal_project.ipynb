```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

df=pd.read_csv('/content/VideoGamesSales.csv')
df
```

```
# df=df[df.duplicated()]
df=df.drop_duplicates()

df['Region']=df['Region'].fillna('North')
#checking null values of region
df[df['Region'].isnull()]
df.head(2)
```

```
print(df.columns)
```

```
df['National Sales']=df['National Sales'].replace('[$]','',regex=True)
df['National Sales'] = pd.to_numeric(df['National Sales'], errors='coerce')
Average_Sales=df['National Sales'].mean()
Average_Sales=int(Average_Sales)
df['National Sales']=df['National Sales'].fillna(Average_Sales)
df.head(2)
```

```
# df=df[df.duplicated()]
df=df.drop_duplicates()

df['Country']=df['Country'].replace({'USA':'United States'})
df['Country']=df['Country'].str.title()
df=df.rename(columns={'NA_Sales':'National Sales','Global_Sales':'Global Sales','NA_Profit':'National Profit','Global_Profit':'Global Profit'})
df

```

```
df=df.rename(columns={'NA_Sales':'National Sales','Global_Sales':'Global Sales','NA_Profit':'National Profit','Global_Profit':'Global Profit'})

sales_cap = df['National Sales'].quantile(0.95)
df['National Sales'] = np.where(df['National Sales'] > sales_cap, sales_cap, df['National Sales'])
df.head(2)
```

```
#create a bar Chart

National_Sales=df.groupby(['Region','Country'])['National Sales'].sum().reset_index().sort_values(by='National Sales', ascending=False)
National_Sales
```

```
plt.figure(figsize=(16, 8))
sns.barplot(data=National_Sales, x='Region', y='National Sales',hue='Country')#hue in this line displays countery bars in different colours
plt.title('National Sales by Region and Country',fontweight='bold')
plt.xlabel('Region',fontweight='bold')
plt.ylabel('National Sales',fontweight='bold')
plt.show()
```

```
#Create Box Plot
# ,meanprops={"marker":"o"
#                                                                                   ,"markerfacecolor":"white"
#                                                                                   ,"markersize":"3"
#                                                                                   ,"markeredgecolor":"green"}
plt.figure(figsize=(10, 6))
sns.boxplot(x='Country', y='National Sales', data=df,showmeans=True,hue='Genre',meanprops={"marker":"o"
                                                                                   ,"markerfacecolor":"white"
                                                                                   ,"markersize":"6"
                                                                                 ,"markeredgecolor":"green"})

plt.title('National Sales by Country',fontweight='bold')
plt.xlabel('Country',fontweight='bold')
plt.ylabel('National Sales',fontweight='bold')
plt.show()
```

```
#create a pie Chart
Sales=df.groupby(['Country'])[['National Sales','Global Sales']].sum().reset_index()
Country=Sales['Country']
National_Sales=Sales['National Sales']
Global_Sales=Sales['Global Sales']

fig, axs = plt.subplots(1, 2, figsize=(10, 7))

axs[0].pie(National_Sales, labels=Country, autopct='%1.1f%%', startangle=90)
axs[0].set_title('National Sales by Country')

axs[1].pie(Global_Sales, labels=Country, autopct='%1.1f%%', startangle=90)
axs[1].set_title('Global Sales by Country')

plt.show()
```

```
# Create a line chart
plt.figure(figsize=(10, 6))

# Line for National Sales
plt.plot(df['Year'], df['National Sales'], marker='o', linestyle='-', color='b', label='National Sales')

# Line for Global Sales
plt.plot(df['Year'], df['Global Sales'], marker='s', linestyle='--', color='g', label='Global Sales')

# Adding titles and labels
plt.title('National and Global Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend()

# Display the line chart
plt.grid(True)
plt.show()
```

This project performs an exploratory data analysis on video game sales data. The steps taken include:


1.  ***Data Loading and Initial Inspection***: The VideoGamesSales.csv dataset was loaded into a pandas DataFrame, and basic information about the data, such as its shape and column types, was inspected.

2.***Data Cleaning and Preparation***
*   Duplicate rows were removed.
*Missing values in the 'Region' column were filled with 'North'.
* The '$' symbol was removed from 'NA_Sales', and the column was converted to a numeric type, with missing values filled with the mean.

* Categorical values in the 'Country' column were standardized (e.g., 'USA' to 'United States') and capitalized.


*  Column names were renamed for clarity ('NA_Sales' to 'National Sales', 'Global_Sales' to 'Global Sales', 'NA_Profit' to 'National Profit', 'Global_Profit' to 'Global Profit').


* Outliers in 'National Sales' were capped at the 95th percentile.

3*** Data Visualization***
*  A bar chart was created to visualize National Sales by Region and Country.

*  A box plot was generated to show National Sales by Country, with different genres highlighted.
*  Pie charts were created to display the proportion of National and Global Sales by Country.


*   A line chart was plotted to show the trend of National and Global Sales over the years.
